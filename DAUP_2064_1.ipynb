Filter the data based on a condition and update the plot to reflect the filtered data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data=pd.read_csv("/content/insurance.csv")
# Filter the data (example: filter for BMI greater than 30)
filtered_data = data[data['age'] > 30]

# Extract x and y values from the filtered data
x_filtered = filtered_data['age']
y_filtered = filtered_data['expenses']

# Create the scatter plot with the filtered data
plt.figure(figsize=(8, 6))
plt.scatter(x_filtered, y_filtered, marker='+', color='red', label='Filtered Data Points (age > 30)')

# Customize the plot (same as before, but now with filtered data)
plt.title('Scatter Plot of age vs. expenses (Filtered)', fontsize=14)
plt.xlabel('age', fontsize=12)
plt.ylabel('expenses', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

data=pd.read_csv("/content/insurance.csv")

age_group_df = data[(data['age'] >= 20) & (data['age'] <= 40)]

# Count the number of smokers in the age group
smoker_counts = age_group_df['smoker'].value_counts()

# Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(age_group_df['age'], age_group_df['smoker'], color='red')
plt.xlabel('Age (20-40)')
plt.ylabel('Smoker (yes/no)')
plt.title('Smokers vs. Age (20-40)')
plt.grid(True)


# Annotate with smoker counts
for i, count in enumerate(smoker_counts):
    plt.annotate(f'{count}', (20, i), textcoords="offset points", xytext=(5, 0), ha='left')

plt.show()
```
